```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from re import sub
from decimal import Decimal
from sklearn.linear_model import LinearRegression   #funções de regressão linear
from sklearn.metrics import mean_squared_error      #função que calcula média de erro quadrático
from sklearn.metrics import mean_absolute_error     #função que calcula média de erro absoluto
from sklearn.model_selection import train_test_split

df = pd.read_csv('usa_mercedes_benz_prices.csv')

def clean(x):
    x = x.replace(" mi.", "").replace(",", "")
    return int(x)
    
df['Mileage'] = df['Mileage'].apply(clean)

df
```

```python
df_util = df[df['Price'] != 'Not Priced']
df_util = df_util[df_util['Mileage'] < 125000] 
df_util
```

```python
quilometragem = df_util.Mileage.values
preco = df_util.Price

preco = [sub[1:] for sub in preco]
preco = [x.replace(',', '.') for x in preco]
preco = [float(x) for x in preco]

fig, ax = plt.subplots()
ax.scatter(quilometragem, preco, s=1)

X_reta1 = quilometragem
y_reta1 = (-3/8000)*X_reta1+50
ax.plot(X_reta1, y_reta1, 'r', label='Hipótese inicial')
plt.legend()                                       #exibe legenda do gráfico

plt.xlabel("Quilometragem")
plt.ylabel("Valor do carro")
plt.show()

err_abs_reta1 = mean_absolute_error(preco, y_reta1)                          #calcula média de erro abs da reta 2
err_qdr_reta1 = mean_squared_error(preco, y_reta1)                           #calcula média de erro quadrático da reta 2
print(f'Erros da reta 1: quadrado = {err_qdr_reta1}, absoluto = {err_abs_reta1}')   #printa erros da reta 2
```

```python
X = quilometragem
y = preco
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
reg = LinearRegression().fit(X_train.reshape(-1, 1), y_train)   #cria e treina um modelo de Regressão Linear com os dados disponíveis
y_pred = reg.predict(X_test.reshape(-1, 1))


err_abs_reta2 = mean_absolute_error(y_test, y_pred)                          #calcula média de erro abs da reta 2
err_qdr_reta2 = mean_squared_error(y_test, y_pred)                           #calcula média de erro quadrático da reta 2
print(f'Erros da reta 2: quadrado = {err_qdr_reta2}, absoluto = {err_abs_reta2}')   #printa erros da reta 2
```

```python
fig, ax = plt.subplots()
ax.scatter(quilometragem, preco, s=1)
ax.plot(X_reta1, y_reta1, 'r', label='Hipótese inicial')

a = reg.coef_[0]                                   #coeficiente angular - função generica do scikit
b = reg.intercept_                                 #coeficiente linear - função genérica do scikit
X_reta2 = quilometragem
y_reta2 = a*X_reta2+b 
ax.plot(X_reta2, y_reta2, 'g', label='Previsão com RL')             #plota reta criada pelo modelo de Regressão Linear

plt.legend()                                       #exibe legenda do gráfico
plt.title('Relação entre quilometragem e preço de venda de carros Mercedes')  #dá um título ao gráfico
plt.xlabel("Quilometragem")                        #legenda eixo x
plt.ylabel("Preço de venda")                     #legenda eixo y
plt.show() 
```
